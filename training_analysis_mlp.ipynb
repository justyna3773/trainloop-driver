```python
from tensorboard.backend.event_processing import event_accumulator
import pandas as pd


def parse_tensorboard(path, scalars):
    """returns a dictionary of pandas dataframes for each requested scalar"""
    ea = event_accumulator.EventAccumulator(
        path,
        size_guidance={event_accumulator.SCALARS: 0},
    )
    _absorb_print = ea.Reload()
    return {k: pd.DataFrame(ea.Scalars(k)) for k in scalars}
```

```python
# Notebook helpers: load a RecurrentPPO model and count parameters

from typing import Union, Optional, Dict, Any
from sb3_contrib import RecurrentPPO
import torch

def _human(n: int) -> str:
    for unit in ["", "K", "M", "B", "T"]:
        if abs(n) < 1000:
            return f"{n:.0f}{unit}"
        n /= 1000.0
    return f"{n:.1f}Q"

def load_recurrentppo(
    model_or_path: Union[str, RecurrentPPO],
    device: str = "cpu",
    custom_objects: Optional[Dict[str, Any]] = None,
) -> RecurrentPPO:
    """
    Load a RecurrentPPO model from path ('.zip') or pass-through if already loaded.
    """
    if isinstance(model_or_path, RecurrentPPO):
        return model_or_path
    # Older/newer SB3 versions differ on load kwargs; try a compatible sequence
    try:
        return RecurrentPPO.load(model_or_path, device=device, custom_objects=custom_objects, print_system_info=False)
    except TypeError:
        return RecurrentPPO.load(model_or_path, device=device, custom_objects=custom_objects)

def count_policy_parameters(
    policy: torch.nn.Module,
    show_breakdown: bool = True,
) -> Dict[str, Any]:
    """
    Count total and trainable parameters of a policy (torch.nn.Module).
    Also returns a top-level submodule breakdown.
    """
    total = sum(p.numel() for p in policy.parameters())
    trainable = sum(p.numel() for p in policy.parameters() if p.requires_grad)

    # Breakdown by top-level module (prefix before the first dot in parameter name)
    breakdown: Dict[str, int] = {}
    for name, p in policy.named_parameters():
        top = name.split(".", 1)[0] if "." in name else name
        breakdown[top] = breakdown.get(top, 0) + p.numel()

    # Pretty print
    print(f"Policy class: {policy.__class__.__name__}")
    print(f"Total parameters:     {total:,}  ({_human(total)})")
    print(f"Trainable parameters: {trainable:,}  ({_human(trainable)})")

    if show_breakdown and breakdown:
        print("\nBreakdown by top-level submodule:")
        width = max(len(k) for k in breakdown)
        for k in sorted(breakdown):
            n = breakdown[k]
            print(f"  {k:<{width}}  {n:,} ({_human(n)})")

    return {
        "policy_class": policy.__class__.__name__,
        "total_params": total,
        "trainable_params": trainable,
        "breakdown": breakdown,
    }

def summarize_recurrentppo(
    model_or_path: Union[str, RecurrentPPO],
    device: str = "cpu",
    custom_objects: Optional[Dict[str, Any]] = None,
    show_breakdown: bool = True,
) -> Dict[str, Any]:
    """
    Convenience wrapper: load (if needed) and summarize.
    """
    model = load_recurrentppo(model_or_path, device=device, custom_objects=custom_objects)
    return count_policy_parameters(model.policy, show_breakdown=show_breakdown)
```

```python
# ---- settings you can tweak ----
SHOW_SEEDS = False        # True = also draw each seed in faint gray
STEP_BIN   = 2_000        # bin width along global_step to align seeds for aggregation
LOW_Q, HI_Q = 0.10, 0.90  # uncertainty band quantiles

def bin_steps(s, width=STEP_BIN):
    return (s // width) * width

# Make every run look the same: must have ['global_step','value']
def normalize_run_df(df, base, run_name):
    df = df.copy()

    # Likely names for step/value if not already present
    step_aliases = [
        'global_step', 'step', 'timestep', 'timesteps', 'n_timesteps',
        'timesteps_total', 'total_timesteps', 'time/total_timesteps'
    ]
    value_aliases = ['value', 'explained_variance', 'eval/explained_variance',
                     'train/explained_variance', 'y']

    # Map step
    if 'global_step' not in df.columns:
        for c in step_aliases:
            if c in df.columns:
                df = df.rename(columns={c: 'global_step'})
                break
        else:
            # last resort: use the index as a monotonic step
            df = df.reset_index(drop=True)
            df['global_step'] = df.index

    # Map value (to_df usually returns 'value', but handle other names)
    if 'value' not in df.columns:
        for c in value_aliases:
            if c in df.columns:
                df = df.rename(columns={c: 'value'})
                break

    # Coerce numerics, drop NaNs, sort
    df['global_step'] = pd.to_numeric(df['global_step'], errors='coerce')
    df['value']       = pd.to_numeric(df['value'],       errors='coerce')
    df = df.dropna(subset=['global_step', 'value']).sort_values('global_step')

    df['chain'] = base
    df['run']   = run_name
    return df

# --------- main ----------
chains = split_into_chains(models)

all_joined_raw = []      # raw per-seed
all_joined_sm  = []      # smoothed per-seed (for aggregation)

fig, ax = plt.subplots(figsize=(10, 6))

for base, names in chains.items():
    run_smooth_dfs = []
    for nm in names:
        try:
            raw = to_df(models[nm], TAG)              # your function
            df  = normalize_run_df(raw, base, nm)     # ensure 'global_step' & 'value'
            all_joined_raw.append(df)

            sm = smooth_ema(df, span=SPAN).copy()     # expects 'global_step','value'; returns 'smooth'
            sm['chain'] = base
            sm['run']   = nm
            sm['step_bin'] = bin_steps(sm['global_step'])
            run_smooth_dfs.append(sm[['chain','run','global_step','step_bin','smooth']])

        except Exception as e:
            print(f"Skipping {nm} due to error: {e}")

    if not run_smooth_dfs:
        continue

    joined_sm = pd.concat(run_smooth_dfs, ignore_index=True)
    all_joined_sm.append(joined_sm)

    # Optional: show each seed faintly
    if SHOW_SEEDS:
        for rn, sub in joined_sm.groupby('run'):
            ax.plot(sub['global_step'], sub['smooth'], lw=0.8, alpha=0.45, color='0.7')

    # Aggregate across seeds on a common (binned) step grid
    agg = (joined_sm
           .groupby(['chain','step_bin'])['smooth']
           .agg(mean='mean',
                low=lambda x: x.quantile(LOW_Q),
                high=lambda x: x.quantile(HI_Q))
           .reset_index()
           .sort_values('step_bin'))

    # Plot mean + uncertainty band
    ax.fill_between(agg['step_bin'], agg['low'], agg['high'], alpha=0.15, linewidth=0)
    ax.plot(agg['step_bin'], agg['mean'], lw=1.8, label=base)

# If you still want the combined frames:
all_joined_raw = pd.concat(all_joined_raw, ignore_index=True) if all_joined_raw else pd.DataFrame()
all_joined_sm  = pd.concat(all_joined_sm,  ignore_index=True) if all_joined_sm  else pd.DataFrame()

# Cosmetics
ax.set_xlabel("Krok czasowy")
ax.set_ylabel("Wyjaniana wariancja (0-1)")
ax.set_title("Wyjaniana wariancja w zalenoci od kroku czasowego (rednia  przedzia)")
ax.set_ylim(0, 1)
ax.grid(True, alpha=0.25)
for spine in ['top','right']:
    ax.spines[spine].set_visible(False)

ax.legend(loc='center left', bbox_to_anchor=(1.02, 0.5), title='acuch')
plt.tight_layout()
plt.show()
```
