### Notebook do liczenia miar podobieństwa zbiorów (Jaccard)

```python
def jaccard(a, b):
    """
    Jaccard index between two iterables of features.
    Returns a float in [0, 1].
    """
    A, B = set(a), set(b)
    if not A and not B:
        return 1.0  # both empty -> maximal overlap by convention
    return len(A & B) / len(A | B)

# Example
features_attention = {0,1,2}
features_spca={0,1,2, 3, 5}
features_ig={0,1,2}

print("Jaccard:", jaccard(features_attention, features_ig))
print("Jaccard:", jaccard(features_spca, features_ig))
```
